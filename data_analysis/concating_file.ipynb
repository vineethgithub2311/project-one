```python
import os
import pandas as pd
```

```python
file_list=os.listdir('data')

df0=pd.DataFrame(pd.read_csv('data/'+file_list[0]))

for file in file_list[1:]:
    df1=pd.DataFrame(pd.read_csv('data/'+file))
    df0=pd.concat([df0,df1])
    
df0.count()
```

```python
df0.to_csv('combined_data.csv')
```
