```python
# Save path to data set in a variable
data_file3 = "community_area_totals.csv"

df_notclean = pd.read_csv(data_file3)
df_community = df_notclean [['pickup_community_area', 'name']]
#df_community.head()

# Save path to data set in a variable
data_file4 = "community_area_totals1.csv"

df_notclean1 = pd.read_csv(data_file4)
df_community1 = df_notclean1 [['dropoff_community_area', 'name']]
df_community1.head()
```

```python
df3 = df1.groupby(['pickup_community_area']).sum()
df3_total_fare = df3[['trip_total']]
df3_total_fare.head()
```

```python
combined_df1 = pd.merge(df_community, df3_total_fare, how='right', on='pickup_community_area')
combined_df1.head()
```

```python
combined_df2 = pd.merge(df2_trips, combined_df1, how='right', on='pickup_community_area')
combined_df2.head()
```

```python
combined_df2['trip_id'] = combined_df2['trip_id'].astype(float)
combined_df2['trip_total'] = combined_df2['trip_total'].astype(float)
combined_df2['trip_average'] = combined_df2['trip_total']/combined_df2['trip_id']
combined_df2.head()
```

```python
df_average_trip = combined_df2.sort_values(by=['trip_average'], ascending=False)
df_average_trip.head()
```

```python
df_average_trip = df_average_trip.dropna(how='any')
df_average_trip
```

```python
df_average_reduced = df_average_trip [['name', 'trip_average']].set_index('name')
df_average_reduced1 = df_average_trip [['name', 'trip_total']].set_index('name')
```

```python
# Create a bar chart
chart1 = df_average_reduced.plot.bar(y='trip_average', rot=90)
chart2 = df_average_reduced1.plot.bar(y='trip_total', rot=90)

# Set the xlabel and ylabel
chart1.set_xlabel("Pick up community area")
chart1.set_ylabel("Trip Average Fare")

chart2.set_xlabel("Pick up community area")
chart2.set_ylabel("Trip Total Fare")

plt.show()
plt.tight_layout()
plt.savefig("../avg_state_rain.png")
```
