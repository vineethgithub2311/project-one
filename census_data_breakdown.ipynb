This notebook uses the census data and the taxi ride data to analyze the taxi ride data

```python
import os 
import pandas as pd
```

```python
files=os.listdir('../analysis_data')
files
```

```python
file_path='../analysis_data/{}'.format(files[5])
```

```python
df=pd.DataFrame(
    pd.read_csv(file_path,
                usecols=['Pickup Census Tract','Trip Start Timestamp','Trip ID'],
               dtype={
                   'Unnamed: 0':int,
                   'Pickup Census Tract':float,
                   'Trip Start Timestamp':object,
                   'Trip ID':int
               }
               )
)
df.head()
```

```python
gb=df.groupby(['Pickup Census Tract','Trip Start Timestamp'])
df2=gb.sum().reset_index()
```

```python
df2.count()
```

```python
df2['Pickup Census Tract']=df2['Pickup Census Tract'].astype('int64')
```

```python
tracts=df2['Pickup Census Tract'].unique()
```

```python
for tract in tracts:
    df3=df2.loc[df2['Pickup Census Tract']==tract]
    df3.to_csv('../analysis_data/census_and_trips/{}.csv'.format(tract))
    print(tract)
```
