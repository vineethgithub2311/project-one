This notebook concats the csv generated from the api requests, cleans and puts it in a new csv file

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_list=os.listdir('data')

df0=pd.DataFrame(pd.read_csv('data/'+file_list[0]))

for file in file_list[1:]:
    df1=pd.DataFrame(pd.read_csv('data/'+file))
    df0=pd.concat([df0,df1])
    
df0.head()
```

```python
df0.columns
```

```python
clean_df=df0[['dropoff_census_tract',
       'dropoff_centroid_latitude',
       'dropoff_centroid_longitude', 'dropoff_community_area',
       'fare', 'payment_type', 'pickup_census_tract',
       'pickup_centroid_latitude',
       'pickup_centroid_longitude', 'pickup_community_area',
              'trip_end_timestamp', 'trip_id', 'trip_miles', 'trip_seconds',
       'trip_start_timestamp', 'trip_total']]
clean_df.to_csv('output/clean_data.csv')
```
