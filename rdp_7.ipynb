```python
import os
import pandas as pd
```

```python
# Store filepath in a variable
taxi_data = "Final_Data/combined_data.csv"
comm_data = "Final_Data/comm_area.csv"
taxi_df = pd.read_csv(taxi_data, encoding="ISO-8859-1")
comm_df = pd.read_csv(comm_data, encoding="ISO-8859-1")
```

```python

comm_df = comm_df.rename({'pickup_community_area':'community_area'},axis=1)
comm_df.columns
```

```python
taxi_fare_df = taxi_df[['trip_id','trip_seconds','trip_miles','fare',
                       'payment_type','tips','trip_total','dropoff_community_area',
                       'pickup_community_area']]
```

```python
tf_drop_df = taxi_fare_df.dropna(how='any')
tf_drop_df.count() 
```

```python
tf_drop_vc = tf_drop_df['dropoff_community_area'].value_counts() 
ht_area_df = pd.DataFrame(tf_drop_vc[tf_drop_vc >= 5000].index, columns=['community_area'])
ht_area_df.columns
```

```python
comb_ht_area = pd.merge(ht_area_df, comm_df, on='community_area',how='inner')
```

```python
#get the area name using droff off. drop columns and rename 

comb_tf_ca_df = pd.merge(tf_drop_df, comb_ht_area, left_on='dropoff_community_area',right_on='community_area',
                         how='inner')
```

```python
comb_tf_ca_df.count() 
```

```python
comb_tf_ca_df.drop('community_area', axis=1, inplace=True)
```
