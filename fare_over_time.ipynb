```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv('../analysis_data/Fare.csv', dtype={'Trip Start Timestamp':object,
                                                    'Fare':float})
df=pd.DataFrame(file)
df.head()
```

```python
df['Trip Start Timestamp']=df['Trip Start Timestamp'].str.split(" ", n = 1, expand = True)
df.head()
```

```python
gb=df.groupby('Trip Start Timestamp')
trip_count=gb.mean().reset_index()
trip_count.head()
```

```python
trip_count['Trip Start Timestamp']=pd.to_datetime(trip_count['Trip Start Timestamp'],yearfirst=True)
```

```python
trip_count['Months']=trip_count['Trip Start Timestamp'].dt.to_period('M')
trip_count.head()
```

```python
gb1=trip_count[['Fare','Months']].groupby('Months')
trip_final=gb1.mean().reset_index()
trip_final=trip_final.drop(index=55)
trip_final
```

```python
fig=trip_final[['Months','Fare']].plot(x='Months',y='Fare',grid=True,
                                                figsize=(15,10),title='Average Fare over Time',
                                                color='r',legend=False, alpha=0.5,)
plt.xlabel('Time')
plt.ylabel('Average Fare ($)')
plt.savefig('output/fare_over_time.png')
```
